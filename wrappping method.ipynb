# Feature Engineering

```python
state1=dataset['State']
state_dummies=pd.get_dummies(state1)
```

```python
state_dummies.head()
```

```python
# Resolving problem of multicolinearity.
```

```python
newstate=state_dummies[['California','Florida']]
```

```python
dataset.columns
```

```python
x=dataset[['R&D Spend', 'Administration', 'Marketing Spend']]
x.head()
```

```python
x[['California','florida']]=newstate
```

```python
x.head()
```

# Feature Extraction

```python
# Wrapping Method
```

```python
y=dataset['Profit']
y
```

```python
import statsmodels.formula.api as sm
```

```python
model_OLS=sm.OLS(endog=y,exog=x).fit()
```

```python
model_OLS.summary()
```
